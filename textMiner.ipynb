```python
import pyspark
from pyspark.sql import SparkSession
from pyspark import SparkContext
```

```python
sc = SparkContext()
```

```python
diario = sc.textFile(r"C:\Users\ramon\Desktop\Dados\diario.txt")
```

```python
diario.collect()
```

```python
palavra = diario.flatMap(lambda x:x.split(""))
```

```python
for p in palavra:
    print(p)
```

```python
def min(x):
    minusc = x.lower()
    return (minusc)
allpalavras = palavra.map(lambda x: min(x))
```

```python
allpalavras.collect()
```

```python
stop_word = ['a', 'de', 'da']
filtraWord = allpalavras.filter(lambda x: x not in stop_word)
```

```python
filtraWord.collect()
```

```python
addColumn = filtraWord.map(lambda x: [x,1])
```

```python
addColumn.collect()
```

```python
from operator import add
```

```python
contWords = addColumn.reduceByKey(add)
```

```python
contWords.collect().show()
```
